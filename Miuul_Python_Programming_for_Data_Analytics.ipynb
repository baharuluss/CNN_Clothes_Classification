#Virtual Environment & Package Management

Pycharm bize hem pythonla çalışmak hemde işletim sistemimizle çalışmak adına alanlar sağlar.Terminali işletim sistemlerimizle konuşmak için kullanırdık.

* Bilgisayarımızdaki sanal ortamların listelenmesi için
  
   * Conda env list komutu kullanılır.
* Kendi sanal ortmımızı oluşturmak
  
 * Conda create -n my_env
  
    Sanal ortam oluşturulmuş oldu
* Sanal Ortamı aktif etme
  
  * conda activate myenv
* Sanal Ortamı deaktif etme
  
  * conda deactivate myenv
* Yüklü Paketlerin Listelenmesi
  
  * conda list
* Paket yükleme:
  
  * conda install numpy
* Aynı amda birden fazla paket yükleme:
  * conda isntall numpy scipy pandas
* Paket silme:
  * conda remove package_name
* Belli bir versiyona göre paket yükleme:
  * conda isntall numpy= 1.20.1
* Paket yükseltme:
  * conda upgrate numpy - en güncel versiyon
* Tüm paketlerin yükseltilmesi:
  * conda upgrade -all

* pip: pypi (python package index) Paket Yönetim Aracı paket yükleme:
  * pip install paket_adi
* Paket Yükleme versiyona göre:
 * pip install pandas== 1.2.1
* Başka çalışma arkadaşına versiyonları aktarmak için:
 * conda env export >environment.yaml
 * conda da yaml file oluşturuyoruz.

* Ortamı silmek için:
 * conda env remove -n my_env
* Ortamı tekrar ayağa kaldırmak için
 * conda env create -f environment.yaml
